## 데이터 전처리

##### 정수형 변수의 결측치는 각 변수별 분산을 고려하여 평균, 중앙값, 최빈값 중 적절한 값으로 대체하였음
##### 데이터 분산에 대한 시각화 결과는 부록에 첨부함.

### 각 컬럼별 분석, 전처리 결과

#### 설립연도

##### 정수형, 결측치 없음. 원본 그대로 사용. 기업 나이 등 파생변수 추가 가능.

#### 국가

##### 코드형 문자열 범주형, 결측치 없음. 원-핫 인코딩 처리.

#### 분야

##### 범주형, 19.5% 결측치 존재. ‘Unknown’ 대체 후 원-핫 인코딩.

#### 투자단계

##### 순서형 범주, 결측치 없음. 숫자 매핑 처리, 모델에 따라 인코딩 조정 가능.

#### 직원 수

##### 실수형, 3.9% 결측치. 평균 대체, 이상치 없음. 필요 시 정수 변환 가능.

#### 인수여부

##### 범주형(예/아니오), 결측치 없음. 원-핫 인코딩 적용.

#### 상장여부

##### 범주형(상장/비상장), 결측치 없음. 원-핫 인코딩 적용.

#### 고객수 (백만명)

##### 실수형, 30.1% 결측치. 중앙값 대체. 로그 변환 및 구간화 가능.

#### 총 투자금 (억원)

##### 실수형, 결측치 없음. 원본 사용, 정수 변환 시 메모리 최적화 가능.

#### 연매출 (억원)

##### 실수형, 결측치 없음. 원본 사용, 정수 변환 시 메모리 최적화 가능.

#### SNS 팔로워 수 (백만명)

##### 수치형(정수 느낌), 결측치 있음. 최빈값 대체. 구간화 분포 특성 존재.

#### 기업가치 (백억원)

##### 구간형 범주, 27.8% 결측치. ‘Unknown’ 또는 평균 추정 대체. 수치화 및 순서형 변수 활용 가능.

### 파생변수 생성

#### 기업나이
##### 기준 연도(2025년)에서 설립연도를 뺀 값으로 기업의 운영 기간을 나타냄.
#### 직원당 매출
##### 연매출(억원)을 직원 수로 나눈 값으로, 1인당 평균 매출 규모를 의미함.
#### 고객당 매출
##### 연매출(억원)을 고객수(백만명)로 나눈 값으로, 고객 1명당 평균 매출 규모를 나타냄
#### 투자수익비
##### 연매출(억원)을 총 투자금*억원(으로 나눈 값으로, 투자 대비 매출 수익성을 나타냄.
#### SNS 팔로워 등급
##### SNS 팔로워 수(백만명)를 기준으로 3단계(낮음: 0~1, 중간: 1~3, 높음: 3~10) 구간화하여 팔로워 규모를 범주형 등급으로 분류
#### 가치대비 매출
##### 기업가치(백억원)를 연매출(억원)으로 나눈 값으로, 기업가치 대비 매출 규모를 평가함
#### 총 투자금당 직원 수
##### 직원 수를 총 투자금(억원)으로 나눈 값으로, 투자금 대비 직원 규모를 나타낸다

#### 스케일링은 Min-Max 스케일링을 적용함

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
import seaborn as sns
import warnings
from sklearn.ensemble import RandomForestRegressor
from IPython.display import Image

# Ignore FutureWarnings
warnings.simplefilter(action='ignore', category=FutureWarning)

rc('font', family='AppleGothic') # 애플 폰트 사용
plt.rcParams['axes.unicode_minus'] = False # 마이너스 부호 안깨지게 설정

test = pd.read_csv('../../data/raw/test.csv')
train = pd.read_csv("../../data/raw/train.csv")
```

### 범주형 변수 고유값

```python
def summarize_categorical(df, columns):
    for col in columns:
        unique_vals = df[col].unique()
        missing_ratio = df[col].isna().mean()
        print(f"### {col}")
        print(f"- 고유값 ({len(unique_vals)}개): {unique_vals}")
        print(f"- 결측치 비율: {missing_ratio:.2%}\n")

# 사용 예
cat_cols = ['국가', '분야', '투자단계', '인수여부', '상장여부']
summarize_categorical(train, cat_cols)
```
