def is_success(row):
    return (
        # ① Series C 이상 투자 유치 여부 (Crunchbase 기준)
        (row['funding_round'] in ['Series C', 'Series D', 'Series E', 'IPO', 'Acquisition'])

        or
        # ② 직원 수가 일정 이상이며 최근 1년간 증가율이 20% 이상 (LinkedIn 기준)
        (row['employee_current'] >= 50 and row['employee_growth_rate'] >= 0.20)

        or
        # ③ 설립 후 5년 이상 생존한 경우 (Crunchbase 기준)
        (row['years_since_founded'] >= 5)

        or
        # ④ 최근 6개월 내 채용 공고가 존재하는 경우 (LinkedIn 기준)
        (row['job_postings_last_6mo'] > 0)
    )

### 크롤링 코드(ex)

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, TimeoutException
from webdriver_manager.chrome import ChromeDriverManager
import time
import csv
from datetime import datetime

# 드라이버 설정
def setup_driver():
    options = webdriver.ChromeOptions()
    options.add_argument("--start-maximized")
    options.add_argument("--disable-blink-features=AutomationControlled")
    options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 Chrome/124.0.0.0 Safari/537.36")
    return webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# 로그인 함수
def login_linkedin(driver, email, password):
    driver.get("https://www.linkedin.com/login")
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID, "username"))).send_keys(email)
    driver.find_element(By.ID, "password").send_keys(password)
    driver.find_element(By.XPATH, "//button[@type='submit']").click()
    WebDriverWait(driver, 10).until(EC.url_contains("/feed"))

# 기업 정보 수집 함수 (About 전용)
def crawl_about_info(driver, company_url):
    about_url = company_url.rstrip("/") + "/about/"
    driver.get(about_url)
    time.sleep(2)

    info = {
        "company_url": company_url,
        "company_name": None,
        "location": None,
        "industry": None,
        "website": None,
        "employee_count": None,
        "company_size": None,
        "founded_year": None,
        "collected_at": datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    }

    # 회사명
    try:
        info["company_name"] = WebDriverWait(driver, 5).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "h1.text-heading-xlarge"))
        ).text
    except:
        pass

    # 위치
    try:
        location_elem = driver.find_element(By.CSS_SELECTOR, "div.org-top-card-summary-info-list__info-item")
        info["location"] = location_elem.text.strip()
    except:
        pass

    # About 상세정보
    try:
        terms = driver.find_elements(By.CSS_SELECTOR, "dt.org-page-details__definition-term")
        texts = driver.find_elements(By.CSS_SELECTOR, "dd.org-page-details__definition-text")
        for term, text in zip(terms, texts):
            title = term.get_attribute("innerText").strip().lower()
            content = text.get_attribute("innerText").strip()

            if "industry" in title:
                info["industry"] = content
            elif "website" in title:
                info["website"] = content
            elif "company size" in title or "회사 규모" in title:
                info["company_size"] = content
                digits = ''.join(filter(str.isdigit, content.split()[0]))
                if digits:
                    info["employee_count"] = int(digits)
            elif "founded" in title or "설립" in title:
                year = ''.join(filter(str.isdigit, content))
                if len(year) == 4:
                    info["founded_year"] = int(year)
    except Exception as e:
        print("About info 수집 실패:", e)

    return info

# 여러 기업 반복 후 저장
def crawl_multiple_companies(email, password, company_urls, output_csv="linkedin_about_data.csv"):
    driver = setup_driver()
    login_linkedin(driver, email, password)

    results = []
    for url in company_urls:
        try:
            data = crawl_about_info(driver, url)
            print(f"Crawled: {data['company_name']}")
            results.append(data)
        except Exception as e:
            print(f"Failed to crawl {url}: {e}")

    driver.quit()

    if results:
        keys = results[0].keys()
        with open(output_csv, "w", newline="", encoding="utf-8-sig") as f:
            writer = csv.DictWriter(f, fieldnames=keys)
            writer.writeheader()
            writer.writerows(results)
        print(f"저장 완료: {output_csv}")
    else:
        print("수집된 정보가 없습니다.")

# 예시 실행
email = "linkedintestabcdefg@gmail.com"
password = "fldzmemdls1234~"

company_urls = [
    "https://www.linkedin.com/company/atommerce/"
]

crawl_multiple_companies(email, password, company_urls)
```

**주요 원인**

LinkedIn은 React 기반 SPA 구조로, 정보가 초기 로딩 시 HTML에 존재하지 않음

사람 눈에는 보이지만, Selenium은 해당 시점에 정보가 없는 HTML을 읽는 상황이 발생함

innerText로 강제로 읽어도, LinkedIn은 로그인한 계정의 상태나 기업 관계 여부에 따라 정보를 숨김

연결되지 않은 계정은 'About' 섹션 일부를 볼 수 없음 (비공개 처리 가능)

LinkedIn은 반자동화 감지 시스템이 매우 강력함

Selenium으로 접속한 브라우저는 일부 정보가 의도적으로 숨겨짐 또는 로딩이 지연됨

법적/약관 측면에서 LinkedIn은 자동화 수집을 제한하고 있음

크롤링 시 IP 차단, 계정 정지, 법적 경고 가능성 존재 (HiQ vs LinkedIn 소송 사례)


**현재 해결을 위해 적용한 방법**:

.text 대신 .get_attribute("innerText")로 값 읽기 → 여전히 값 없음

WebDriverWait으로 로딩 대기 적용 → 일부 요소는 여전히 비어 있음

About 페이지 직접 URL 접근 시도 → 페이지 접근은 가능하나 데이터는 없음

로그인을 직접 입력하는 수동 방식 적용 → 여전히 비어 있음
